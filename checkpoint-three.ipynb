## Irregular Data

Detect outliers in your dataset and handle them as needed. Use code comments to make notes about your thought process.

```python
import numpy as np

# Thought Process
# Check if 'pairing_quality' falls outside the valid 0-5 scale using describe().
# Use np.where() to find any scores < 0 or > 5 (outliers) and drop them to ensure data integrity.

# Step 1: Statistical Check
print(df['pairing_quality'].describe())

# Step 2: Detect scores outside the 0-5 range
outliers = np.where((df['pairing_quality'] < 0.0) | (df['pairing_quality'] > 5.0))[0]

# Step 3: Remove outliers if they exist
if len(outliers) > 0:
    df = df.drop(df.index[outliers])
    print(f"Removed {len(outliers)} rows with invalid scores.")
else:
    print("No numerical outliers detected.")
```
